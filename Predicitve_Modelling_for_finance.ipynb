```python
### TWO YER RETURNS OF BEML



df_beml = yf.download("BEML.NS", start="2004-01-01", end="2024-01-01")
df_beml.reset_index(inplace=True)


df_beml['Year'] = df_beml['Date'].dt.year

df_beml['Two_Year'] = (df_beml['Year'] // 2) * 2


two_year_data = df_beml.groupby('Two_Year')['Close'].agg(['first', 'last']) #grouping by year and taking first and last


two_year_data['Two_Year_Returns'] = (two_year_data['last'] - two_year_data['first']) / two_year_data['first'] * 100


all_year = pd.DataFrame({'Two_Year_Group': np.arange(2004, 2025, 2)})
two_year_data = two_year_data.reindex(all_year['Two_Year_Group']).fillna(0)



colors = ['green' if x >= 0 else 'red' for x in two_year_data['Two_Year_Returns']]


plt.figure(figsize=(25, 10))
plt.bar(two_year_data.index, two_year_data['Two_Year_Returns'], color=colors , width = 1.8)
plt.xlabel('Year', fontsize=34,color = "red")
plt.ylabel('2-Year Returns (%)', fontsize=34, color = "darkblue")
plt.title('2-Year Stock Returns of BEML (2004 - 2024)', fontsize=38, color = "black")

plt.xticks(two_year_data.index, fontsize=28, rotation=45)
plt.yticks(fontsize=28)
plt.grid()
```

### LINEAR REGRESSION BEML


```python
df.columns

```

```python
#Taking close as an dependent

X = df[["Open", "High", "Low" ,"Volume", "Relative_Strength_Index", "50_Day_Simple_Moving_Average" ,"200_Day_Simple_Moving_Average"]]
y = df[["Close"]]
```

```python
from sklearn.model_selection import train_test_split , cross_val_score

from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_squared_error , mean_absolute_error

from sklearn.metrics import r2_score , make_scorer
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)


LR_Model = LinearRegression()

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

LR_Model.fit(X_train, y_train)


y_predict = LR_Model.predict(X_test)



mse_= mean_squared_error(y_test, y_predict)
mae_ = mean_absolute_error(y_test, y_predict)

rmse_ = np.sqrt(mse_)
r2_ = r2_score(y_test, y_predict)



### we are doing it for TEST SET
print("Before Cross Validation MSE of BEML", mse_)
print("Before Cross Validation RMSE of BEML" , rmse_)
print("Before Cross Validation MAE of BEML", mae_)

print("Before Cross Validation R2 score BEML is " , r2_)



#CROSS VALIDATION

mse_after_cv= cross_val_score(LR_Model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
mae_after_cv = cross_val_score(LR_Model, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')
r2_after_cv = cross_val_score(LR_Model, X_train, y_train, cv=5, scoring= "r2")


##converting the scores to positive by negative errors.
mse_score = -mse_after_cv
mae_score = -mae_after_cv


print("----------------------------------------------------")


print("After Cross-Validation the mean MSE of BEML ", mse_score.mean())

print("After Cross-Validation the mean MAE of BEML ", mae_score.mean())

print("After Cross-Validation the mean R Score of BEML:", r2_after_cv.mean())
```

```python
from sklearn.svm import SVR
```

```python


X_train, X_test, y_train, y_test = train_test_split(X_svr, y_svr, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Initialize and train the SVR model
svr_model_beml = SVR(kernel="linear", gamma=0.1)
svr_model_beml.fit(X_train, y_train)

#predictions on the test set
y_predict = svr_model_beml.predict(X_test)

#performance metrics on the test set
svr_mse_beml = mean_squared_error(y_test, y_predict)
svr_mae_beml = mean_absolute_error(y_test, y_predict)
svr_rmse_beml = np.sqrt(svr_mse_beml)
r2_svr = r2_score(y_test, y_predict)

#performance metrics before cross-validation
print("Before Cross-Validation (SVR) MSE of BEML:", svr_mse_beml)
print("Before Cross-Validation (SVR) RMSE of BEML:", svr_rmse_beml)
print("Before Cross-Validation (SVR) MAE of BEML:", svr_mae_beml)
print("Before Cross-Validation (SVR) R of BEML:", r2_svr)

print("--------------------------------------------------------")

#Perform cross-validation
mse_after_cv_beml = cross_val_score(svr_model_beml, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
mae_after_cv_beml = cross_val_score(svr_model_beml, X_train, y_train, cv=5, scoring='neg_mean_absolute_error')
r2_after_cv_beml = cross_val_score(svr_model_beml, X_train, y_train, cv=5, scoring="r2")

# Convert the scores to positive values for MSE and MAE
mse_score_beml = -mse_after_cv_beml
mae_score_beml = -mae_after_cv_beml

# Print the cross-validation results
print("After Cross-Validation the mean MSE of BEML:", mse_score_beml.mean())
print("After Cross-Validation the mean MAE of BEML:", mae_score_beml.mean())
print("After Cross-Validation the mean R Score of BEML:", r2_after_cv_beml.mean())
```
